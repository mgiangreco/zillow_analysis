```python
import pandas as pd
import numpy as np
from datetime import datetime
import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots

pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 50)
```

```python
METADATA_COLS = ['RegionID', 'RegionName', 'City', 'State', 'Metro', 'CountyName', 'SizeRank']
```

See https://www.zillow.com/research/data/

```python
zri = pd.read_csv('/Users/markgiangreco/Documents/data/zillow/Neighborhood_Zri_AllHomesPlusMultifamily.csv')

zhvi = pd.read_csv('/Users/markgiangreco/Documents/data/zillow/Neighborhood_Zhvi_AllHomes.csv')
```

See https://www.zillow.com/research/2020-hot-markets-south-26293/

```python
OUTPERFORM_CITIES = ['Atlanta', 'Austin', 'Charlotte', 'Nashville', 'Tampa'
         , 'Dallas', 'Denver', 'San Antonio', 'Jacksonville', 'Houston']

LARGEST_CITIES = zri[zri['SizeRank']<=10].City.unique()

OTHER_CITIES = ['Salt Lake City']
```

```python
zri.head()
```

```python
zhvi.head()
```

```python
zri_log = zri

for col in zri_log.columns.difference(METADATA_COLS):
    zri_log[col] = np.log(zri_log[col])
```

```python
zhvi_log = zhvi

for col in zhvi_log.columns.difference(METADATA_COLS):
    zhvi_log[col] = np.log(zhvi_log[col])
```

```python
rent = zri.groupby('City')[(zri.columns.difference(METADATA_COLS))].mean()\
.T.reset_index().rename(columns={'index': 'Month'})

rent.head()
```

```python
rent_log = zri_log.groupby('City')[(zri_log.columns.difference(METADATA_COLS))].mean()\
.T.reset_index().rename(columns={'index': 'Month'})

rent_log.head()
```

```python
homevalue = zhvi.groupby('City')[(zhvi.columns.difference(METADATA_COLS))].mean()\
.T.reset_index().rename(columns={'index': 'Month'})

homevalue.head()
```

```python
homevalue_log = zhvi_log.groupby('City')[(zhvi_log.columns.difference(METADATA_COLS))].mean()\
.T.reset_index().rename(columns={'index': 'Month'})

homevalue_log.head()
```

```python
# plot log-rents and log-prices of cities of interest

for i in list(set(OUTPERFORM_CITIES) | set(LARGEST_CITIES) | set(OTHER_CITIES)):
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    fig.add_trace(go.Scatter(
                x=rent['Month'],
                y=rent[i],
                name='rent',
                line_color='deepskyblue',
                opacity=0.8), 
                secondary_y=False,)
    
    fig.add_trace(go.Scatter(
                x=homevalue['Month'],
                y=homevalue[i],
                name='home value',
                line_color='dimgray',
                opacity=0.8),
                secondary_y=True,)
    
    fig.update_layout(title_text='Average Rent & Home Values, {}'.format(i), 
                     xaxis_range=['2010-09','2020-01'])

    fig.show()
    
    
```

```python
# plot log-rents and log-prices of cities of interest

for i in list(set(OUTPERFORM_CITIES) | set(LARGEST_CITIES) | set(OTHER_CITIES)):
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    fig.add_trace(go.Scatter(
                x=rent_log['Month'],
                y=rent_log[i],
                name='log(rent)',
                line_color='deepskyblue',
                opacity=0.8), 
                secondary_y=False,)
    
    fig.add_trace(go.Scatter(
                x=homevalue_log['Month'],
                y=homevalue_log[i],
                name='log(home value)',
                line_color='dimgray',
                opacity=0.8),
                secondary_y=True,)
    
    fig.update_layout(title_text='Average Log(Rent) & Log(Home Values), {}'.format(i), 
                     xaxis_range=['2010-09','2020-01'])

    fig.show()
    
```
